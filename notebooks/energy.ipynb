```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/energy/summary.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

```

```python
df["Unit/Clutch"] = pd.to_numeric(df["Unit/Clutch"], errors="coerce")
df["Diesel Stock in litres"] = pd.to_numeric(df["Diesel Stock in litres"], errors="coerce")
```

```python
(df.isna().sum() / len(df)) * 100
```

```python
df = df.dropna(subset=["Clutch Produced"])
```

```python
df.drop(columns=["Unnamed: 17"], inplace=True)
```

```python
df = df.sort_values("Date").reset_index(drop=True)
```

```python
df.fillna({"Diesel_per_Clutch": 0}, inplace=True)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df['Date'], df['Daily Energy consumption in units'])
plt.xlabel('Date')
plt.ylabel('Energy (Units)')
plt.title('Daily Energy Consumption Trend')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.plot(df['Date'], df['To date power cost in INR'], label='Electricity Cost')
plt.plot(df['Date'], df['Cumulative Diesel cost in INR'], label='Diesel Cost')
plt.xlabel('Date')
plt.ylabel('Cumulative Cost (INR)')
plt.title('Cumulative Power vs Diesel Cost')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.scatter(df['Clutch Produced'], df['Daily Energy consumption in units'])
plt.xlabel('Clutch Produced')
plt.ylabel('Energy Consumption (Units)')
plt.title('Production vs Energy Consumption')
plt.tight_layout()
plt.show()
```

```python
df.to_csv("preprocessed_power_data.csv", index=False)
```
